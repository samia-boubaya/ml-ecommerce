````python
import pandas as pd

# Load main dataset
df = pd.read_csv("../data/raw/online_shoppers_intention.csv")

# Define the Renaming Dictionary
rename_columns = {
    "Administrative": "admin",
    "Administrative_Duration": "admin_duration",
    "Informational": "info",
    "Informational_Duration": "info_duration",
    "ProductRelated": "prod_related",
    "ProductRelated_Duration": "prod_related_duration",
    "BounceRates": "bounce_rate",
    "ExitRates": "exit_rate",
    "PageValues": "page_value",
    "SpecialDay": "special_day",
    "Month": "month",
    "OperatingSystems": "os",
    "Browser": "browser",
    "Region": "region",
    "TrafficType": "traffic_type",
    "VisitorType": "visitor_type",
    "Weekend": "weekend",
    "Revenue": "revenue"
}

# Rename columns using the rename Dictionary
df.rename(columns=rename_columns, inplace=True)

# Save updated dataset into a new CSV file
df.to_csv("../data/processed/online_shoppers_intention_01_standard.csv", index=False)

# Display Standardized DataFrame
print("Standardized DataFrame:")
display(df.head())

# Display the list of the standardidized column names in df
print("Standardized Column Names in df:")
display(df.info())
````
