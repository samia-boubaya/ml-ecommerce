# Hypothesis Testing

```python
import pandas as pd
import numpy as np
from statsmodels.stats.proportion import proportions_ztest
```

```python
df = pd.read_csv('../data/processed/online_shoppers_intention_06_clean.csv')
df.head()
```

## Two-proportion Z-test for Weekend vs Weekday conversions

- H0: Conversion rate (weekday) = Conversion rate (weekend)
- H1: Conversion rates are different

Significance Value - 0.05

```python
# Create weekday and weekend subsets
weekday = df[df["weekend"] == False]
weekend = df[df["weekend"] == True]

# Number of conversions
conv_weekday = weekday["revenue"].sum()
conv_weekend = weekend["revenue"].sum()

# Number of observations
n_weekday = len(weekday)
n_weekend = len(weekend)

# Z-test
count = np.array([conv_weekday, conv_weekend])
nobs = np.array([n_weekday, n_weekend])

z_stat, p_value = proportions_ztest(count, nobs)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

- The p-value is less than significance value. Null Hypothesis is rejected, hence there is a statistically significant difference.

- The difference in conversion rates between weekdays and weekends is statistically significant, indicating that weekend visitors have a meaningfully higher likelihood of purchasing compared to weekday visitors

## Independent two sample T-test

- H0: The mean Page Value is the same for converting and non-converting sessions
- H1: The mean Page Value differs between converting and non-converting sessions

Significance Value - 0.05

```python
from scipy.stats import ttest_ind

# Split Page Value by Revenue class
pagevalue_no = df[df["revenue"] == 0]["page_value"]
pagevalue_yes = df[df["revenue"] == 1]["page_value"]

# Perform t-test (Welch's t-test: unequal variances safer)
t_stat, p_value = ttest_ind(pagevalue_yes, pagevalue_no, equal_var=False)

print("Mean Page Value (No Purchase):", pagevalue_no.mean())
print("Mean Page Value (Purchase):", pagevalue_yes.mean())
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

- The p-value is extremely less than significance value. Null Hypothesis is rejected, hence converting sessions have a substantially higher average Page Value than non-converting sessions
