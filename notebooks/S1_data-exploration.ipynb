```python
import pandas as pd
from IPython.display import display

# Load Raw data online_shoppers_intention.csv into a DataFrame df_main 
df_main = pd.read_csv("../data/raw/online_shoppers_intention.csv")

# Display df_main DataFrame
print("Online Shoppers Intention Dataset:")
display(df_main.head(10))
print("Main Dataset Shape:", df_main.shape[0], "rows and", df_main.shape[1], "columns\n")

# Display data types & missing values of each column in df_main
print("Data Types & Missing Values of Each Column:")
display(df_main.info())

# Loop through each column and print unique values
for col in df_main.columns:
    unique_vals = df_main
[col].unique()
    print(f"Column: {col}")
    print(f"Unique values ({len(unique_vals)}): {unique_vals}\n")

print(f"\nData types check:")
for col in df_main.columns:
    dtype = df_main
[col].dtype
    unique_vals = df_main
[col].nunique()
    print(f"  ‚û°Ô∏è{col:<25} {str(dtype):<10} [{unique_vals}] unique values")
```
